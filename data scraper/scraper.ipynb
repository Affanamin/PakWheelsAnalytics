```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
#car_name = "toyota-corolla"
#base_url = f"https://www.pakwheels.com/used-cars/{car_name}/688?page="

#car_name = "vitz"
#base_url = f"https://www.pakwheels.com/used-cars/search/-/?q={car_name}&page="

#car_name = "honda-civic"
#car_code = "642"

#"kia-sportage","kia-picanto",
#"715","51368",

car_names = ["kia-stonic","kia-sorento","kia-classic",
             "haval-h6","haval-jolion",
             "mg-hs--2","mg-zs","mg-zs-ev","mg-u9--2",
             "changan-oshan-x7","changan-alsvin","changan-karvaan",
             "jac-t9","jac-x200"
            ]
car_codes = ["777495","660713","666",
             "721544","714593",
             "612997","635218","590258","3087057",
             "844463","641668","313798",
             "2867543","183425"
            ]


#base_url = f"https://www.pakwheels.com/used-cars/{car_name}/{car_code}?page="
```

```python

chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--ignore-certificate-errors")
driver = webdriver.Chrome(options=chrome_options)


feature_list = ["Alloy Wheels","Front Fog Lights","DRLs","Sun Roof","ABS","Air Bags","Air Conditioning","Alloy Rims",
"Infotainment System","Front Speakers", "Rear Speakers", "Steering Switches", "Paddle Shifters",
"Traction Control","Front Camera", "Rear Camera", "Immobilizer Key", "Power Locks", "Power Steering",
"Cool Box","Cruise Control","Climate Control"]

#csvFilePath = f"scraped_data_{str(car_name)}.csv"
```

```python
#csvFilePath
```

```python
import csv
for car_name, car_code in zip(car_names, car_codes):
    
    base_url = f"https://www.pakwheels.com/used-cars/{car_name}/{car_code}?page="
    csvFilePath = f"scraped_data_{car_name}.csv"

    with open(csvFilePath, mode='w', newline ='', encoding='utf-8') as csvfile:
        csv_writer = csv.writer(csvfile)
        headerRow = ["MakeModel", "Year", "Mileage", "FuelType", "Transmission", "City", "Colour", "ManufactureType", 
                     "EngineCapacity", "CarType", "PostLastUpdatedat", "CarAddress", "Demand"] + feature_list
        csv_writer.writerow(headerRow)
        
        for page_number in range (1,20):
            url = base_url + str(page_number)
            driver.get(url)
        
            driver.implicitly_wait(20)
            
            html_content=driver.page_source
            soup = BeautifulSoup(html_content, 'html.parser')
        
            ul_class = "list-unstyled search-results search-results-mid next-prev car-search-results"
            div_child_class = "col-md-9 grid-style"
            anchor_tag_class = "car-name ad-detail-path"
            
            concat_urls = []
        
            ul_list = soup.find_all('ul',{'class':ul_class})
            if ul_list:
                for ul_element in ul_list:
                    specific_div_list = ul_element.find_all('div', {'class': div_child_class})
                    if specific_div_list:
                        for specific_div in specific_div_list:
                            anchor_tags = specific_div.find_all('a',{'class':anchor_tag_class})
                            if anchor_tags:
                                for tag in anchor_tags:
                                    hrefTag = tag.get("href")
                                    if hrefTag:
                                        #print(hrefTag)
                                        fullUrl = urljoin("https://www.pakwheels.com", hrefTag)
                                        concat_urls.extend([fullUrl])
                                        #print(concat_urls)
                                        
                                        #concat_lll = concat_urls.extend([urljoin("https://www.pakwheels.com/",hrefTag)])
                                        
                            else:
                                print("NMothing Fouind")
                    else:
                        print("all messed up")
                    
                    for index, link in enumerate(concat_urls, start=1):
                        driver.get(link)
                        driver.implicitly_wait(10)
                        inner_html_content = driver.page_source
                        inner_soup = BeautifulSoup(inner_html_content, 'html.parser')
                        #print(inner_soup)
        
                        h1_inner_html = []
                        table_td_inner_html = []
                        odd_li_inner_html = []
                        detail_sub_heading_inner_html = ""
                        generic_green_tag_inner_html = ""
        
                        try:
                            h1_tag = inner_soup.find('div',{'class':'well'}).find('h1').text.strip()
                            h1_inner_html.append(h1_tag)
                            #print(h1_inner_html)
                        except AttributeError:
                            print("Could not find <h1> tag inside <div class='well'>")
        
                        try:
                            table_tag = inner_soup.find('table', {'class': 'table table-bordered text-center table-engine-detail fs16'})
                            if table_tag:
                                tds = []
                                for tr_tag in table_tag.find_all('tr'):
                                    for td_tag in tr_tag.find_all('td'):
                                        tds.append(td_tag.text.strip())
                                table_td_inner_html.extend(tds)
                                #print(table_td_inner_html)
                        except AttributeError:
                            print("Could not find <table> with class=table table-bordered text-center table-engine-detail fs16 ")
        
                        try:
                            ul_featured = inner_soup.find('ul',{'class': 'list-unstyled ul-featured clearfix'})
                            if ul_featured:
                                odd_li_inner_html.extend([li_tag.text.strip() for i, li_tag in enumerate(ul_featured.find_all('li')) if i % 2 != 0])
                                #print(odd_li_inner_html)
                        except AttributeError:
                            print("Could not find <ul> with class='list-unstyled ul-featured clearfix")
        
                        try:
                            ul_tag = inner_soup.find('ul',{'class': 'list-unstyled car-feature-list nomargin'})
                            if ul_tag:
                                ul_inner_html = ul_tag.decode_contents()
                                comparision_results = [1 if feature in ul_tag.stripped_strings else 0 for feature in feature_list]
                                #print(f"Comaprisioin Results are,{comparision_results}")
                        except AttributeError:
                            print("Could not find <ul> with class=list-unstyled car-feature-list nomargin")
        
                        try:
                            detail_sub_heading_tag = inner_soup.find('p',{'class':'detail-sub-heading'})
                            if detail_sub_heading_tag:
                                a_tag = detail_sub_heading_tag.find('a')
                                if a_tag:
                                    detail_sub_heading_inner_html = a_tag.text.strip()
                                    #print(f"detail_sub_heading_inner_html are: {detail_sub_heading_inner_html}")
                        except AttributeError:
                            print("Could not find <p class = detail-sub-heading")
        
                        try:
                            generic_green_tag = inner_soup.find('strong', {'class': 'generic-green'})
                            if generic_green_tag:
                                generic_green_tag_inner_html = generic_green_tag.text.strip()
                                #print(f"generic_green_tag_inner_html are: {generic_green_tag_inner_html}")
                                #print(generic_green_tag_inner_html)
                        except AttributeError:
                            print("Could Not Find <strong> tag with in 'class': 'generic-green' ")
    
    
                        
                        #comparision_results = [1 if feature in ul_tag.stripped_strings else 0 for feature in feature_list]
                        csv_writer.writerow(h1_inner_html + table_td_inner_html + odd_li_inner_html + [detail_sub_heading_inner_html]+ [generic_green_tag_inner_html] + comparision_results)
                        
                        print(f"loop number {str(page_number)}")

driver.quit()

print(f"ScrapedData saved to {csvFilePath}")
                            
                
               
```
