```python
##GOld
from pyspark.sql import functions as F
from pyspark.sql.types import *
from pyspark.sql.functions import monotonically_increasing_id
from delta.tables import DeltaTable
from pyspark.sql.functions import current_timestamp, lit
spark.conf.set("spark.sql.session.timeZone", "UTC")
```

```python
#base_silver_path = dbutils.fs.ls("abfss://silver@staccpakwheels.dfs.core.windows.net/pakwheels/cardata")
base_silver_path = "abfss://silver@staccpakwheels.dfs.core.windows.net/pakwheels/cardata"
#display(dbutils.fs.ls(base_silver_path[0].path))
```

```python
silver_df = spark.read \
    .format("delta") \
    .load(base_silver_path)
```

```python
silver_df.count()
```

```python
## checking data

from pyspark.sql import functions as F
from pyspark.sql.window import Window


non_postid_cols = [c for c in silver_df.columns if c != "postId"]


w = Window.partitionBy(*non_postid_cols).orderBy(F.lit(1))

dedup_df = (
    silver_df
    .withColumn("rn", F.row_number().over(w))
    .filter(F.col("rn") == 1)
    .drop("rn")
)



print("Before:", silver_df.count())
print("After :", dedup_df.count())


#silver_df = silver_df[[~dup_groups]]
```

```python
dedup_df.show(2,False)
```

```python
feature_cols = ["AlloyWheels","FrontFogLights","DRLs","SunRoof", "ABS","AirBags","AirConditioning","AlloyRims","InfotainmentSystem","FrontSpeakers","RearSpeakers","SteeringSwitches","PaddleShifters","TractionControl","FrontCamera","RearCamera","ImmobilizerKey","PowerLocks","PowerSteering","CoolBox","CruiseControl", "SafetyScore", "ComfortScore", "TechScore"]

dedup_df = dedup_df.withColumn(
    "feature_hash",
    F.sha2(F.concat_ws("||", *feature_cols), 256)
)
```

```python
## creating hash for Car Dimension

dedup_df = dedup_df.withColumn(
    "car_hash",
    F.sha2(F.concat_ws("||", "Brand","MakeModel","Year","EngineCapacity","FuelType","Transmission","CarType","Colour"), 256),
)
```

```python
## creating hash for Date Dimension

dedup_df = dedup_df.withColumn(
    "date_hash",
    F.sha2(F.concat_ws("||", "PostLastUpdatedat","PostYear","PostMonth","PostDayOfWeek"), 256),
)
```

```python
## creating hash for location dim

dedup_df = dedup_df.withColumn(
    "location_hash",
    F.sha2(F.concat_ws("||", "City","CityGroup","CarAddress"), 256),
)
```

```python
dim_car = dedup_df.select(
    "car_hash","Brand","MakeModel","Year","EngineCapacity",
    "FuelType","Transmission","CarType","ManufactureType"
).dropDuplicates() \
.withColumn("car_id", monotonically_increasing_id())
```

```python
### Car_Dim
targetPathCarDim = "abfss://gold@staccpakwheels.dfs.core.windows.net/goodDataFolder/dimenstions/dim_car"

if DeltaTable.isDeltaTable(spark, targetPathCarDim):
    
    delta_table = DeltaTable.forPath(spark, targetPathCarDim)

    delta_table.alias("t").merge(
        dim_car.alias("s"),
        "t.MakeModel = s.MakeModel AND t.is_current = true"
    ) \
    .whenMatchedUpdate(
        condition="""
            t.Brand != s.Brand OR
            t.EngineCapacity != s.EngineCapacity
        """,
        set={
            "effective_end_date": "current_timestamp()",
            "is_current": "false"
        }
    ) \
    .whenNotMatchedInsert(values={
        "car_id": "s.car_id",
        "Brand": "s.Brand",
        "MakeModel": "s.MakeModel",
        "Year": "s.Year",
        "EngineCapacity": "s.EngineCapacity",
        "FuelType": "s.FuelType",
        "Transmission": "s.Transmission",
        "CarType": "s.CarType",
        "ManufactureType": "s.ManufactureType",
        "effective_start_date": "current_timestamp()",
        "effective_end_date": "null",
        "is_current": "true"
    }).execute()

else:
    dim_car.withColumn("effective_start_date", current_timestamp()) \
           .withColumn("effective_end_date", lit(None)) \
           .withColumn("is_current", lit(True)) \
           .write.format("delta").save(targetPathCarDim)
```

```python
### location_dim
dim_location = dedup_df.select(
    "location_hash","City","CityGroup","CarAddress"
).dropDuplicates() \
.withColumn("location_id", monotonically_increasing_id())

targetPathLocationDim = "abfss://gold@staccpakwheels.dfs.core.windows.net/goodDataFolder/dimenstions/dim_location"

if DeltaTable.isDeltaTable(spark, targetPathLocationDim):
    
    delta_table = DeltaTable.forPath(spark, targetPathLocationDim)

    delta_table.alias("t").merge(
        dim_location.alias("s"),
        "t.location_id = s.location_id AND t.is_current = true"
    ) \
    .whenMatchedUpdate(
        condition="""
            t.CityGroup != s.CityGroup OR
            t.CarAddress != s.CarAddress
        """,
        set={
            "effective_end_date": "current_timestamp()",
            "is_current": "false"
        }
    ) \
    .whenNotMatchedInsert(values={
        "location_id": "s.location_id",
        "City": "s.City",
        "CityGroup": "s.CityGroup",
        "CarAddress": "s.CarAddress",
        "effective_start_date": "current_timestamp()",
        "effective_end_date": "null",
        "is_current": "true"
    }).execute()

else:
    dim_location.withColumn("effective_start_date", current_timestamp()) \
           .withColumn("effective_end_date", lit(None)) \
           .withColumn("is_current", lit(True)) \
           .write.format("delta").save(targetPathLocationDim)


```

```python
## date_dim
dim_date = dedup_df.select(
    "date_hash","PostLastUpdatedat","PostYear","PostMonth","PostDayOfWeek"
).dropDuplicates() \
.withColumn("date_id", monotonically_increasing_id())

targetPathDateDim = "abfss://gold@staccpakwheels.dfs.core.windows.net/goodDataFolder/dimenstions/dim_date"

if DeltaTable.isDeltaTable(spark, targetPathDateDim):
    
    delta_table = DeltaTable.forPath(spark, targetPathDateDim)

    delta_table.alias("t").merge(
        dim_date.alias("s"),
        "t.date_id = s.date_id AND t.is_current = true"
    ) \
    .whenMatchedUpdate(
        condition="""
            t.date_hash != s.date_hash OR
            t.PostMonth != s.PostMonth OR
            t.PostDayOfWeek != s.PostDayOfWeek
        """,
        set={
            "effective_end_date": current_timestamp(),
            "is_current": False
        }
    ) \
    .whenNotMatchedInsertAll() \
    .execute()
else:
    dim_date.write.format("delta").save(targetPathDateDim)


```

```python
### car_features_dim
dim_car_features = dedup_df.select(
    "AlloyWheels","FrontFogLights","DRLs","SunRoof", "ABS","AirBags","AirConditioning","AlloyRims","InfotainmentSystem","FrontSpeakers","RearSpeakers","SteeringSwitches","PaddleShifters","TractionControl","FrontCamera","RearCamera","ImmobilizerKey","PowerLocks","PowerSteering","CoolBox","CruiseControl", "SafetyScore", "ComfortScore", "TechScore"
).dropDuplicates() \
.withColumn("car_features_id", monotonically_increasing_id())

feature_cols = ["AlloyWheels","FrontFogLights","DRLs","SunRoof", "ABS","AirBags","AirConditioning","AlloyRims","InfotainmentSystem","FrontSpeakers","RearSpeakers","SteeringSwitches","PaddleShifters","TractionControl","FrontCamera","RearCamera","ImmobilizerKey","PowerLocks","PowerSteering","CoolBox","CruiseControl", "SafetyScore", "ComfortScore", "TechScore"]

dim_car_features = dim_car_features.withColumn(
    "feature_hash",
    F.sha2(F.concat_ws("||", *feature_cols), 256)
)


targetPathCarFeaturesDim = "abfss://gold@staccpakwheels.dfs.core.windows.net/goodDataFolder/dimenstions/dim_car_features"

if DeltaTable.isDeltaTable(spark, targetPathCarFeaturesDim):
    
    delta_table = DeltaTable.forPath(spark, targetPathCarFeaturesDim)

    delta_table.alias("t").merge(
        dim_car_features.alias("s"),
        "t.AlloyWheels = s.AlloyWheels AND t.is_current = true"
    ) \
    .whenMatchedUpdate(
        condition="""
            t.FrontFogLights != s.FrontFogLights OR
            t.DRLs != s.DRLs OR
            t.SunRoof != s.SunRoof OR
            t.ABS != s.ABS OR
            t.AirBags != s.AirBags OR
            t.AirConditioning != s.AirConditioning OR
            t.AlloyRims != s.AlloyRims OR
            t.InfotainmentSystem != s.InfotainmentSystem OR
            t.FrontSpeakers != s.FrontSpeakers OR
            t.RearSpeakers != s.RearSpeakers OR
            t.SteeringSwitches != s.SteeringSwitches OR
            t.PaddleShifters != s.PaddleShifters OR
            t.TractionControl != s.TractionControl OR
            t.FrontCamera != s.FrontCamera OR
            t.RearCamera != s.RearCamera OR
            t.ImmobilizerKey != s.ImmobilizerKey OR
            t.PowerLocks != s.PowerLocks OR
            t.PowerSteering != s.PowerSteering OR
            t.CoolBox != s.CoolBox OR
            t.CruiseControl != s.CruiseControl
        """,
        set={
            "effective_end_date": current_timestamp(),
            "is_current": False
        }
    ) \
    .whenNotMatchedInsertAll() \
    .execute()
else:
    dim_car_features.write.mode("overwrite").save(targetPathCarFeaturesDim)
    


```

```python
## Fact Table:::

# listing_id (PK) → postId
# car_id (FK)
# location_id (FK)
# date_id (FK)
# feature_id (FK)

# -- Measures
# DemandPKR
# DemandInLacs
# Mileage
# MileagePerYear
# CarAge
# ListingAgeDays

# -- Derived
# PriceCategory
# MileageBucket
# MissingCriticalInfo


targetPathCarFeaturesFact = "abfss://gold@staccpakwheels.dfs.core.windows.net/goodDataFolder/facts/facts_car_listing"
fact_df = dedup_df \
    .join(dim_car, ["car_hash"], "left") \
    .join(dim_location, ["location_hash"], "left") \
    .join(dim_date, ["date_hash"], "left") \
    .join(dim_car_features, ["feature_hash"] , "left") \
    .select(
        F.col("postId").alias("listing_id"),
        F.col("car_id").alias("car_dim_id"),
        F.col("location_id").alias("location_dim_id"),
        F.col("date_id").alias("date_dim_id"),
        F.col("car_features_id").alias("car_features_dim_id"),
        "DemandPKR",
        "DemandInLacs",
        "Mileage",
        "MileagePerYear",
        "CarAge",
        "ListingAgeDays",
        "PriceCategory",
        "MileageBucket",
        "MissingCriticalInfo"
    )


fact_df.write \
    .format("delta") \
    .mode("append") \
    .save(targetPathCarFeaturesFact)
```
